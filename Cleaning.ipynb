```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
building_permits_orig = pd.read_csv('data/building_permits.csv')
code_violations_orig = pd.read_csv('data/code_complaints_violations.csv')
```

```python
code_violations = code_violations_orig
building_permits = building_permits_orig
```

```python
complaint_source = code_violations.groupby('Source of Complaint').count()["Number"]
complaint_source
```

```python
#Percent of total complaints that are proactive
complaint_source.Proactive / complaint_source.sum()
```

```python
#Number of Cases handled by each inspector
code_violations.groupby('Inspector').count().sort_values('Number', ascending=False)["Parcel Number"]
```

```python
#Freq of each violation type

violation_freq = code_violations.groupby("Case Type").count().sort_values('Number', ascending=False)["Number"]
violation_freq
# violation_freq.plot()
```

```python
#replace invalid parcel numbers with 0
code_violations['Parcel Number'] = code_violations['Parcel Number'].replace('- -', 0 )
code_violations['Parcel Number'] = code_violations['Parcel Number'].replace('00000- 000- 000', 0 )
```
