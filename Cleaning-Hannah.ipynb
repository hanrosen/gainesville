```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

%matplotlib inline
```

```python
building_permits_orig = pd.read_csv('data/building_permits.csv')
code_violations_orig = pd.read_csv('data/code_complaints_violations_2011_19.csv')
```

```python
code_violations = code_violations_orig
building_permits = building_permits_orig
```

```python
code_violations.head()
```

```python
building_permits.head()
```

# Cleaning code violations dataset

```python
code_violations.shape
```

```python
#replace invalid parcel numbers with 0
code_violations['Parcel Number'] = code_violations['Parcel Number'].replace('- -', 0 )
code_violations['Parcel Number'] = code_violations['Parcel Number'].replace('00000- 000- 000', 0 )
code_violations['Parcel Number'] = code_violations['Parcel Number'].replace('99999- 999- 999', 0 )
```

```python
#convert dates to datetime
code_violations["Compliance Date"] = pd.to_datetime(code_violations["Compliance Date"])
code_violations["Compliance Due Date"] = pd.to_datetime(code_violations["Compliance Due Date"])
code_violations["Infraction Date"] = pd.to_datetime(code_violations["Infraction Date"])
```

```python
code_violations.head()
code_violations.shape
```

```python
# code_violations=code_violations.replace('0',np.nan).dropna(axis = 0, how = 'any')
# new_code_violations = code_violations[code_violations['Parcel Number'] != 0]
new_code_violations = code_violations[code_violations['Parcel Number'] != 0]
new_code_violations
```

# Cleaning building permits dataset

```python
building_permits.head()
```

```python
building_permits["Submit"] = pd.to_datetime(building_permits["Submit"])
building_permits["Issue"] = pd.to_datetime(building_permits["Issue"])
```

# Analysis Code Complaints

## Source of Complaints

```python
complaint_source = code_violations.groupby('Source of Complaint').count()["Number"]
complaint_source
```

```python
#Percent of total complaints that are proactive
complaint_source.Proactive / complaint_source.sum()
```

## Violations per Inspector

```python
#Number of Cases handled by each inspector
code_violations.groupby('Inspector').count().sort_values('Number', ascending=False)["Parcel Number"]
```

## Frequency of each violation type

```python
#Freq of each violation type

violation_freq = code_violations.groupby("Case Type").count().sort_values('Number', ascending=False)["Number"]
violation_freq
# violation_freq.plot()
```

## Violations per zip code

```python
#sorted by zip code
code_violations.groupby("Florida Census Zip Codes").count().sort_values("Florida Census Zip Codes", ascending = True)["Number"]
```

```python
#sorted by count
code_violations.groupby("Florida Census Zip Codes").count().sort_values("Number", ascending = False)["Number"]
```

## Violations handled by Police District

```python
code_violations.groupby("Gainesville Police Districts").count().sort_values("Number", ascending = False)["Number"]
```

## Case Status

```python
case_status = code_violations.groupby("Case Status").count().sort_values("Number", ascending = False)["Number"]
case_status
```
